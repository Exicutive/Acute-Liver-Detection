Defining the Data Augmentation Pipeline

```python
# Load the dataset from a single directory
data_dir = '/content/drive/MyDrive/Liver'

# Remove any unwanted hidden files or directories
# For example, to remove `.ipynb_checkpoints`
cleaned_dir = [f for f in os.listdir(data_dir) if not f.startswith('.')]
print(cleaned_dir)

dataset= image_dataset_from_directory(
    data_dir,
    labels='inferred',

    # Keep batching for efficiency
    batch_size = 32,

    label_mode='int',
    interpolation='nearest',
    shuffle=True,

    # Set a dummy image size to bypass initial resizing
    image_size=(1, 1))

# Check the dataset
for images, labels in dataset.take(1):
    print(images.shape, labels.shape)
```

1. Data Preprocessing

    * Normalization: Normalize the pixel values of the images (usually to a range of [0, 1]) to standardize the input data.

    * Data Augmentation: If your dataset is small, apply data augmentation techniques like rotation, flipping, and zooming to artificially increase the diversity of the training data.

    * Splitting the Data: Split the dataset into training, validation, and test sets. A common split is 70% for training, 15% for validation, and 15% for testing.

I manually splitting the dataset using tf.data.experimental.cardinality and skip/take functions. This ensures that the splits are accurately managed.

```python
import numpy as np
import os
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import RandomFlip, RandomRotation, RandomZoom


# Get list of all files and corresponding labels
normal_dir = os.path.join(data_dir, 'Normal')
abnormal_dir = os.path.join(data_dir, 'Abnormal')

normal_files = [os.path.join(normal_dir, f) for f in os.listdir(normal_dir)]
abnormal_files = [os.path.join(abnormal_dir, f) for f in os.listdir(abnormal_dir)]

all_files = normal_files + abnormal_files
all_labels = [0] * len(normal_files) + [1] * len(abnormal_files)

# Shuffle and split the dataset
files_train, files_rem, labels_train, labels_rem = train_test_split(all_files, all_labels, train_size=0.7, stratify=all_labels)
files_val, files_test, labels_val, labels_test = train_test_split(files_rem, labels_rem, train_size=0.5, stratify=labels_rem)

# Define the augmentation pipeline
data_augmentation = tf.keras.Sequential([
    RandomFlip('horizontal'),
    RandomRotation(0.2),
    RandomZoom(0.2)
])

# Create tensorflow datasets
def load_preprocess_and_augment(image_file, label, augment=False):
    image = tf.io.read_file(image_file)
    image = tf.image.decode_jpeg(image, channels=3)

    image = tf.image.resize(image, [224, 224])

    # Normalize to [0, 1]
    image = image / 255.0

    # Apply data augmentation if specified (only for training)
    if augment:
        image = data_augmentation(image)

    # Ensure consistent output shape after augmentation
    image = tf.image.resize_with_crop_or_pad(image, 224, 224)

    return image, label

def create_dataset(file_list, label_list, augment=False):
    dataset = tf.data.Dataset.from_tensor_slices((file_list, label_list))
    dataset = dataset.map(lambda x, y: load_preprocess_and_augment(x, y, augment), num_parallel_calls=tf.data.experimental.AUTOTUNE)
    dataset = dataset.batch(32).prefetch(buffer_size=tf.data.experimental.AUTOTUNE)
    return dataset

# Create the training dataset with augmentation
train_dataset = create_dataset(files_train, labels_train, augment=True)


validation_dataset = create_dataset(files_val, labels_val, augment=False)
test_dataset = create_dataset(files_test, labels_test, augment=False)
```

```python
import matplotlib.pyplot as plt

def display_images(dataset, num_images=5):

    # Taking one batch from the dataset
    for images, labels in dataset.take(1):

        # Adjusting figure size for better visibility
        plt.figure(figsize=(15, 15))

        # Ensure we don't exceed the batch size
        for i in range(min(num_images, images.shape[0])):
            ax = plt.subplot(1, num_images, i + 1)

            # Rescale the image for display
            img = images[i].numpy()

            # Converting back to [0, 255] range for display
            img = (img * 255).astype("uint8")

            plt.imshow(img)
            plt.title(f"Label: {int(labels[i])}")
            plt.axis("off")
        plt.show()

# Displaying images from the training dataset
display_images(train_dataset, num_images=5)
```

Modeling Architecture Selection

When working with a small dataset, it's crucial to choose a model architecture that can effectively leverage the limited data available while avoiding overfitting.

I will be using Transfer Learning with Pre-trained Model (VGG16)

We add custom layers on top of the base model for classification.

We compile the model with an appropriate loss function and optimizer.



```python
from tensorflow.keras.applications import VGG16
from tensorflow.keras import layers, models
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam

# Load the VGG16 model with pretrained weights
base_model = VGG16(input_shape=(224, 224, 3), include_top=False, weights='imagenet')

inputs = layers.Input(shape=(224, 224, 3), name = "input_layer")

# Add custom layers to the base model
x = base_model(inputs)
x = GlobalAveragePooling2D()(x)
x = layers.Dense(64, activation='relu')(x)

# For binary classification (normal vs abnormal)
outputs = Dense(1, activation='sigmoid')(x)

# Compile Model
vgg16_model = Model(inputs, outputs)
vgg16_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001), loss='binary_crossentropy', metrics=['accuracy'])
```

Training the Model

Use the training and validation datasets to train the model. Monitor metrics like accuracy and loss on both training and validation data to prevent overfitting.

The model is trained using the fit method, which involves iterating over the data in batches and updating the model weights to minimize the loss function.

```python
vgg16_model_history = vgg16_model.fit(train_dataset,
                                      steps_per_epoch=len(train_dataset),
                                      epochs = 20,
                                      validation_data=test_dataset,
                                      validation_steps = len(test_dataset))
```


Evaluating the Model on the test Dataset

```python
test_loss, test_accuracy = vgg16_model.evaluate(test_dataset)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```

```python
def plot_history(VGG16_history, model_name):
    plt.figure(figsize=(12, 6))

    # Plot training & validation accuracy values
    plt.subplot(1, 2, 1)
    plt.plot(VGG16_history.history['accuracy'])
    plt.plot(VGG16_history.history['val_accuracy'])
    plt.title(f'{model_name} Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(['Train', 'Validation'])

    # Plot training & validation loss values
    plt.subplot(1, 2, 2)
    plt.plot(VGG16_history.history['loss'])
    plt.plot(VGG16_history.history['val_loss'])
    plt.title(f'{model_name} Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(['Train', 'Validation'])

    plt.show()

plot_history(VGG16_history, 'VGG16')
```

Using Convolution Neural Network

The CNN model is constructed using the TensorFlow and Keras libraries. The architecture comprises several convolutional layers, pooling layers, and fully connected layers. This configuration enables the model to learn and extract features at various levels of abstraction.

```python
from tensorflow.keras import layers, models
CNN_model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(filters = 64,
                            kernel_size = 3,
                            input_shape = (224, 224, 3),
                            activation = "relu"),

    # First Convolutional Layer
    tf.keras.layers.Conv2D(64, 3, activation='relu'),
    tf.keras.layers.MaxPool2D(2),

    # Second Convolutional Layer
    tf.keras.layers.Conv2D(128, 3, activation='relu'),
    tf.keras.layers.Conv2D(128, 3, activation='relu'),
    tf.keras.layers.MaxPool2D(2),

    # Third Convolutional Layer
    tf.keras.layers.Conv2D(256, 3, activation='relu'),
    tf.keras.layers.Conv2D(256, 3, activation='relu'),
    tf.keras.layers.MaxPool2D(2),

    # Flatten the output to feed into the Dense layers
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(1, activation='sigmoid'),

])

CNN_model.compile(loss='binary_crossentropy',
              optimizer=tf.keras.optimizers.Adam(),
              metrics=['accuracy'])
```

Convolution Neural Network Training

The model was trained using the fit method from TensorFlow's Keras API. The training process was configured to run for 20 epochs, with the progress monitored using both training and validation datasets.

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='val_loss', patience=3)
CNN_history = CNN_model.fit(train_dataset,
                            epochs = 50,
                            steps_per_epoch = len(train_dataset),
                            validation_data = test_dataset,
                            validation_steps = len(test_dataset),
                            callbacks=[early_stopping])
```

```python
test_loss, test_accuracy = CNN_model.evaluate(test_dataset)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```

```python
def plot_history(CNN_history, model_name):
    plt.figure(figsize=(12, 6))

    # Plot training & validation accuracy values
    plt.subplot(1, 2, 1)
    plt.plot(CNN_history.history['accuracy'])
    plt.plot(CNN_history.history['val_accuracy'])
    plt.title(f'{model_name} Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(['Train', 'Validation'])

    # Plot training & validation loss values
    plt.subplot(1, 2, 2)
    plt.plot(CNN_history.history['loss'])
    plt.plot(CNN_history.history['val_loss'])
    plt.title(f'{model_name} Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(['Train', 'Validation'])

    plt.show()

plot_history(CNN_history, 'CNN')
```
